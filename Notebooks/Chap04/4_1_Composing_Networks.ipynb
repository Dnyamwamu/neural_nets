```python
# Imports math library
import numpy as np
# Imports plotting library
import matplotlib.pyplot as plt
```

```python
# Define the Rectified Linear Unit (ReLU) function
def ReLU(preactivation):
  activation = preactivation.clip(0.0)
  return activation
```

```python
# Define a shallow neural network with, one input, one output, and three hidden units
def shallow_1_1_3(x, activation_fn, phi_0,phi_1,phi_2,phi_3, theta_10, theta_11, theta_20, theta_21, theta_30, theta_31):
  # Initial lines
  pre_1 = theta_10 + theta_11 * x
  pre_2 = theta_20 + theta_21 * x
  pre_3 = theta_30 + theta_31 * x
  # Activation functions
  act_1 = activation_fn(pre_1)
  act_2 = activation_fn(pre_2)
  act_3 = activation_fn(pre_3)
  # Weight activations
  w_act_1 = phi_1 * act_1
  w_act_2 = phi_2 * act_2
  w_act_3 = phi_3 * act_3
  # Combine weighted activation and add y offset
  y = phi_0 + w_act_1 + w_act_2 + w_act_3
  # Return everything we have calculated
  return y
```

```python
# # Plot two shallow neural networks and the composition of the two
def plot_neural_two_components(x_in, net1_out, net2_out, net12_out=None):

  # Plot the two networks separately
  fig, ax = plt.subplots(1,2)
  fig.set_size_inches(8.5, 8.5)
  fig.tight_layout(pad=3.0)
  ax[0].plot(x_in, net1_out,'r-')
  ax[0].set_xlabel('Net 1 input'); ax[0].set_ylabel('Net 1 output')
  ax[0].set_xlim([-1,1]);ax[0].set_ylim([-1,1])
  ax[0].set_aspect(1.0)
  ax[1].plot(x_in, net2_out,'b-')
  ax[1].set_xlabel('Net 2 input'); ax[1].set_ylabel('Net 2 output')
  ax[1].set_xlim([-1,1]);ax[1].set_ylim([-1,1])
  ax[1].set_aspect(1.0)
  plt.show()

  if net12_out is not None:
    # Plot their composition
    fig, ax = plt.subplots()
    ax.plot(x_in ,net12_out,'g-')
    ax.set_xlabel('Net 1 Input'); ax.set_ylabel('Net 2 Output')
    ax.set_xlim([-1,1]);ax.set_ylim([-1,1])
    ax.set_aspect(1.0)
    plt.show()
```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

# Define ReLU activation
def ReLU(preactivation):
    return preactivation.clip(0.0)

# Define a shallow 1-1-3 neural network
def shallow_1_1_3(x, activation_fn, phi_0, phi_1, phi_2, phi_3,
                   theta_10, theta_11, theta_20, theta_21, theta_30, theta_31):
    # Linear preactivations
    pre_1 = theta_10 + theta_11 * x
    pre_2 = theta_20 + theta_21 * x
    pre_3 = theta_30 + theta_31 * x

    # Activations
    act_1 = activation_fn(pre_1)
    act_2 = activation_fn(pre_2)
    act_3 = activation_fn(pre_3)

    # Weighted activations
    w_act_1 = phi_1 * act_1
    w_act_2 = phi_2 * act_2
    w_act_3 = phi_3 * act_3

    # Output
    y = phi_0 + w_act_1 + w_act_2 + w_act_3
    return y

# Function to plot networks
def plot_neural_two_components(x_in, net1_out, net2_out, net12_out=None):
    fig, ax = plt.subplots(1,2, figsize=(12,5))
    ax[0].plot(x_in, net1_out,'r-'); ax[0].set_title("Net 1 Output")
    ax[0].set_xlim([-1,1]); ax[0].set_ylim([-1,1])
    ax[1].plot(x_in, net2_out,'b-'); ax[1].set_title("Net 2 Output")
    ax[1].set_xlim([-1,1]); ax[1].set_ylim([-1,1])
    plt.show()

    if net12_out is not None:
        plt.figure(figsize=(6,5))
        plt.plot(x_in, net12_out,'g-'); plt.title("Composed Network Output")
        plt.xlim([-1,1]); plt.ylim([-1,1])
        plt.show()

# Input range
x = np.arange(-1,1,0.001)

# First network parameters
n1_theta_10, n1_theta_11 = 0.0, -1.0
n1_theta_20, n1_theta_21 = 0.0, 1.0
n1_theta_30, n1_theta_31 = -0.67, 1.0
n1_phi_0, n1_phi_1, n1_phi_2, n1_phi_3 = 1.0, -2.0, -3.0, 9.3

# Second network parameters
n2_theta_10, n2_theta_11 = -0.6, -1.0
n2_theta_20, n2_theta_21 = 0.2, 1.0
n2_theta_30, n2_theta_31 = -0.5, 1.0
n2_phi_0, n2_phi_1, n2_phi_2, n2_phi_3 = 0.5, -1.0, -1.5, 2.0

# Run first and second networks
net1_out = shallow_1_1_3(x, ReLU, n1_phi_0, n1_phi_1, n1_phi_2, n1_phi_3,
                          n1_theta_10, n1_theta_11, n1_theta_20, n1_theta_21, n1_theta_30, n1_theta_31)
net2_out = shallow_1_1_3(x, ReLU, n2_phi_0, n2_phi_1, n2_phi_2, n2_phi_3,
                          n2_theta_10, n2_theta_11, n2_theta_20, n2_theta_21, n2_theta_30, n2_theta_31)

# Plot individual networks
plot_neural_two_components(x, net1_out, net2_out)

# Compose networks (feed Net 1 output into Net 2)
net12_out = shallow_1_1_3(net1_out, ReLU, n2_phi_0, n2_phi_1, n2_phi_2, n2_phi_3,
                           n2_theta_10, n2_theta_11, n2_theta_20, n2_theta_21, n2_theta_30, n2_theta_31)

# Plot composed network
plot_neural_two_components(x, net1_out, net2_out, net12_out)

# Modify first network and compose again
net1_out_mod = shallow_1_1_3(x, ReLU, n1_phi_0, n1_phi_1*0.5, n1_phi_2, n1_phi_3,
                              n1_theta_10, n1_theta_11, n1_theta_20, n1_theta_21, n1_theta_30, n1_theta_31)
net12_out_mod = shallow_1_1_3(net1_out_mod, ReLU, n2_phi_0, n2_phi_1, n2_phi_2, n2_phi_3,
                               n2_theta_10, n2_theta_11, n2_theta_20, n2_theta_21, n2_theta_30, n2_theta_31)

# Plot modified composition
plot_neural_two_components(x, net1_out_mod, net2_out, net12_out_mod)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of linear regions in the first (red) and second (blue) networks
R = 7   # Example: 3 hidden units, 2D input
S = 5   # Example: second network

# Maximum number of stacked copies of the first network
N_max = 6

# Compute total number of regions for 1 to N_max copies
linear_regions = []
for N in range(1, N_max+1):
    total_regions = R**N * S
    linear_regions.append(total_regions)

# Plot
plt.figure(figsize=(8,5))
plt.plot(range(1, N_max+1), linear_regions, 'o-', linewidth=2)
plt.yscale('log')  # Exponential growth -> log scale
plt.xlabel('Number of stacked copies of first network (N)')
plt.ylabel('Total linear regions (approx.)')
plt.title('Exponential growth of linear regions with depth')
plt.grid(True)
plt.show()
```
